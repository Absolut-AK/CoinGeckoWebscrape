```python
from bs4 import BeautifulSoup
import cloudscraper
```

```python
url = 'https://www.coingecko.com/'
#need to have cloudscraper because web is blocking requests
scrape = cloudscraper.create_scraper()
html = scrape.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
#getting the column of the table
column = soup.find('thead')
column = column.find_all('th')
column = [i.text.strip() for i in column][1:-1]
```

```python
import pandas as pd

df = pd.DataFrame(columns=column)
```

```python
#getting Row of table
table = soup.find('tbody', {'data-target':'currencies.contentBox'})
row = table.find_all('tr')
for i in row:
    lists = [j.text.strip('\n') for j in i][1:-1]
    #cleaning the data a bit
    while('' in lists):
        lists.remove('')
    df.loc[len(df)] = lists
print(df.to_string())
```
